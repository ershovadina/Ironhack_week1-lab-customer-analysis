```python
import pandas as pd
```

```python
# 1.Read the three files into python as dataframes
```

```python
file1 = pd.read_csv('csv_files/file1.csv')
file1.head()
```

```python
file2 = pd.read_csv('csv_files/file2.csv')
file2.head()
```

```python
file3 = pd.read_csv('csv_files/file3.csv')
file3.head()
```

```python
# 2. Show the DataFrame's shape.
```

```python
print(file1.shape)
print(file2.shape)
print(file3.shape)
```

```python
# 3. Standardize header names. (use lower case, if headers have spaces, replace them with underscores)
```

```python
headers1 = file1.columns
headers2 = file2.columns
headers3 = file3.columns

print(headers1)
print(headers2)
print(headers3)
```

```python
cols1 = []
for i in range (len(headers1)):
    header_new = headers1[i].replace(" ", "_")
    cols1.append(header_new.lower())

cols2 = []
for i in range (len(headers2)):
    header_new = headers2[i].replace(" ", "_")
    cols2.append(header_new.lower())

cols3 = []
for i in range (len(headers3)):
    header_new = headers3[i].replace(" ", "_")
    cols3.append(header_new.lower())


print(cols1)
print(cols2)
print(cols3)
```

```python
file1.columns = cols1
file2.columns = cols2
file3.columns = cols3

cols1
```
